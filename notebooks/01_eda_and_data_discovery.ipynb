# EchoChain Notebook 01: Exploratory Data Analysis & Discovery
This notebook inspects the raw datasets (Manufacturing BOM, Warranty Claims, Marketplace Listings, and User E-Waste dataset).

```python
import pandas as pd
import json

# Load Manufacturing BOM
df_bom = pd.read_csv('../data/raw/manufacturing_bom.csv')
print('BOM Records:', len(df_bom))
df_bom.head()
```

```python
# Load Warranty Claims
df_warranty = pd.read_csv('../data/raw/warranty_claims.csv')
print('Warranty Claims Records:', len(df_warranty))
df_warranty.head()
```

```python
# Load Marketplace Listings
with open('../data/raw/marketplace_listings.json', 'r', encoding='utf-8') as f:
    listings = json.load(f)
df_listings = pd.DataFrame(listings)
print('Secondary Marketplace Listings:', len(df_listings))
df_listings.head()
```
